```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("Downloads/diabetes.csv")
data
```

```python
y = data.Outcome
X = data.drop('Outcome', axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid = train_test_split(X,y,train_size = 0.8,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(C = 100, penalty = 'l2', solver = 'liblinear')
```

```python
model.fit(X_train,y_train)
y_pred = model.predict(X_valid)
```

```python
from sklearn.metrics import mean_absolute_error,accuracy_score,classification_report,confusion_matrix
print(confusion_matrix(y_pred, y_valid))
print(classification_report(y_pred, y_valid))
print(accuracy_score(y_valid,y_pred))
print(mean_absolute_error(y_pred, y_valid))
```

```python
para = {
    'C': [100,10,1,0.1,0.01],
    'penalty': ['none','l1','l2','elasticnet'],
    'solver': ['newton-cg','lbfgs','liblinear','sag','saga']
}
```

```python
from sklearn.model_selection import GridSearchCV
best = GridSearchCV(LogisticRegression(),para)
best.fit(X_train,y_train)
```

```python
best.best_params_
```

```python
import evalml
from evalml.preprocessing import split_data
X_train, X_valid, y_train, y_valid = split_data(X, y, problem_type='binary')
from evalml.automl import AutoMLSearch
model = AutoMLSearch(X_train=X_train, y_train=y_train, problem_type='binary')
model.search()
model.rankings
```

```python
model.describe_pipeline(model.rankings.iloc[1]['id'])
```
